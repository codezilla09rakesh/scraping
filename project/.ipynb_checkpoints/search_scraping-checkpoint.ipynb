```python
import requests as req
from bs4 import BeautifulSoup as bs
import re
```

```python
url = "https://www.amazon.in/s?k=ac"
user_agent = "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.96 Safari/537.36"
accept_language = "en"
headers = {
    "User-Agent":user_agent,
    "Accept-Language":accept_language
}
html_response = req.get(url, headers=headers)
```

```python
soup = bs(html_response.text, "lxml")
# print(soup.prettify())
```

```python
page_length = soup.find('li', attrs={"aria-disabled":"true"})
page_length.text
```

```python
s=soup.find_all("div", attrs={"data-component-type":"s-search-result"})
print(len(s))
count = 1
for i in s:
    img = i.find("img")['src']
    title = i.find("h2")
    product_link = "https://www.amazon.in"+title.find('a')['href']
    rating = i.find("span", attrs={"class":"a-size-base","dir":"auto"})
    price = i.find("span", class_="a-price-whole")
    print(count)
    print('img=====>',img)
    try:
        print("rating",rating.text)
    except:
        pass
    print("price", price.text)  
    print("product Link", product_link)
    print()
    print("-------",title.text)
    count +=1
```
