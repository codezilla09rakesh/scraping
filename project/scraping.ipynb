```python
import requests as req
from bs4 import BeautifulSoup as bs
import re
```

```python
url = "https://www.amazon.in/ask/questions/asin/B08GXC2NTX/2/"
user_agent = "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.96 Safari/537.36"
accept_language = "en"
headers = {
    "User-Agent":user_agent,
    "Accept-Language":accept_language
}
html_response = req.get(url, headers=headers)
```

```python
soup = bs(html_response.text, "lxml")
# print(soup.prettify())
```

```python
ques = soup.find(class_="askTeaserQuestions")
# geting all question
# que = ques.find_all(class_="a-fixed-left-grid a-spacing-small")
# for i in range(10):
#     r = que[i].find_all("span")
#     # for q in que:
#     #     print(q.text.strip(""))
#     regex = re.compile(r'[\n\r\t]')
#     # r[0].text
#     q=r[1].text
#     print(regex.sub(" ",q).strip(" "))
print(ques)
```

```python
s=ques.find_all(class_='a-fixed-left-grid-col a-col-right')
counter = 1
for i in range(10):
    r=s[i*3].find_all(class_="a-fixed-left-grid-col a-col-right")
    que = r[0].text
    regex = re.compile(r'[\n\r\t]')
    que=regex.sub("",que).strip(" ")
    print("Que : ",que)
    counter +=1
    e=r[1].find("span")
    g=e.find(class_="askLongText")
    if g:
        ans=g.text
        ans=regex.sub("",ans).strip(" ")
        print("Ans :  ", ans)
    else:
        ans=e.text
        ans=regex.sub("",ans).strip(" ")
        print("Ans :  ", ans)
    print()
```
